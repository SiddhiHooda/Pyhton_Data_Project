```python
import pandas as pd
import matplotlib.pyplot as plt

#IMPORTING DATASET
from datasets import load_dataset
dataset=load_dataset('lukebarousse/data_jobs') 
df=dataset['train'].to_pandas()

#CLEANING THE DATA
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
import ast
def clean_string(skill_list):
    if pd.notna(skill_list):
        return ast.literal_eval(skill_list)
    else:
        return skill_list
df['job_skills']=df['job_skills'].apply(clean_string)
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie', startangle=90, autopct='%1.1f')
plt.ylabel('')
plt.title("Work From home Status")
```

```python
df[['job_work_from_home', 'job_no_degree_mention', 'job_health_insurance']]
```

```python
fig, ax=plt.subplots(1,3)
dict_columns={
    'job_work_from_home':'Work from Home',
    'job_no_degree_mention':'Degree Req.',
    'job_health_insurance': 'Health insurance'
}
for i,(column,title) in enumerate(dict_columns.items()):
    df[column].value_counts().plot(kind='pie',ax=ax[i],startangle=90, autopct='%1.1f')
    ax[i].set_ylabel('')
    ax[i].set_title(title)
#fig (figure object): The entire figure that contains all subplots.
#ax (axes object or array of subplots): A list (array) 
```
